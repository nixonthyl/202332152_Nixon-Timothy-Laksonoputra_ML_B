```python
#IMPORT LIBRARY
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("botak.csv")
```

## BAR CHART

```python
#Bar Chart: Rata-rata probabilitas botak berdasarkan pendidikan
plt.figure(figsize=(10, 6))
edu_avg = df.groupby("pendidikan")['botak_prob'].mean().sort_values()
sns.barplot(x=edu_avg.index, y=edu_avg.values, palette="coolwarm")
plt.xlabel("Tingkat Pendidikan")
plt.ylabel("Rata-rata Probabilitas Botak")
plt.title("Rata-rata Probabilitas Botak Berdasarkan Pendidikan")
plt.xticks(rotation=45)
plt.show()
```

## LINE CHART

```python
#Line Chart: Tren probabilitas botak berdasarkan usia
plt.figure(figsize=(10, 6))
age_avg = df.groupby("umur")['botak_prob'].mean()
plt.plot(age_avg.index, age_avg.values, marker='o', linestyle='-', color='b')
plt.xlabel("Usia")
plt.ylabel("Rata-rata Probabilitas Botak")
plt.title("Tren Probabilitas Botak Berdasarkan Usia")
plt.grid()
plt.show()
```

## PIE CHART

```python
#Pie Chart: Distribusi penggunaan sampo
plt.figure(figsize=(8, 8))
shampoo_counts = df['sampo'].value_counts()
plt.pie(shampoo_counts, labels=shampoo_counts.index, autopct='%1.1f%%', colors=sns.color_palette("pastel"))
plt.title("Distribusi Penggunaan Sampo")
plt.show()
```
